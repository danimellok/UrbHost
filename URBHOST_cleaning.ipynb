```python
# Calculate the length of each 'House Rules' entry
data['House Rules Length'] = data['House Rules'].fillna('').apply(len)

# Calculate the average length of 'House Rules'
average_length = data['House Rules Length'].mean()
print(f"Average length of 'House Rules': {average_length:.2f}")

# Categorize 'House Rules' based on length
def categorize_house_rules(length, avg_length):
    if length < avg_length:
        return 'Weak'
    elif length == avg_length:
        return 'Average'
    else:
        return 'Strict'

data['House Rules Category'] = data['House Rules Length'].apply(lambda x: categorize_house_rules(x, average_length))

# Display the counts of each category
category_counts = data['House Rules Category'].value_counts()
print("Counts for 'House Rules Category':")
print(category_counts)
```

```python
# Define the threshold for categorization
length_threshold = 170.10

# Fill missing 'House Rules Category' based on the threshold
data['House Rules Category'] = data['House Rules Length'].apply(
    lambda x: 'Weak' if x <= length_threshold else 'Strict'
)

# Display the counts for the updated categories
category_counts = data['House Rules Category'].value_counts()
print("Counts for 'House Rules Category':")
print(category_counts)
```
